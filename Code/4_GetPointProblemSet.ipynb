```python
import requests
import pickle
import json
import time

# 从文件中加载 Cookies
def load_cookies_from_file(cookies_file_path='cookies.pkl'):
    with open(cookies_file_path, 'rb') as f:
        cookies = pickle.load(f)
    return cookies

# 使用保存的 Cookies 发送 API 请求
def request_with_cookies(api_url, cookies_file_path='cookies.pkl'):
    # 加载 Cookies
    cookies = load_cookies_from_file(cookies_file_path)

    # 创建 session，使用保存的 cookies
    session = requests.Session()

    # 将 cookies 添加到 session
    for cookie in cookies:
        session.cookies.set(cookie['name'], cookie['value'])

    # 设置请求头
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Accept': 'application/json;charset=UTF-8',
        'Referer': 'https://pintia.cn/problems?knowledgePointId=63',
        'Content-Type': 'application/json;charset=UTF-8'
    }

    # 发送请求
    response = session.get(api_url, headers=headers)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"请求失败，状态码: {response.status_code}")
        return None

# 递增页码抓取题目信息并添加重试机制，单个题目出错时跳过
def fetch_problems(page=0, limit=30, cookies_file_path='cookies.pkl', retries=3):
    api_url = f"https://pintia.cn/api/problems?page={page}&limit={limit}&filter=%7B%22statuses%22%3A%5B%22PENDING_REVIEW%22%2C%22REVIEWED%22%2C%22REJECTED%22%5D%2C%22knowledgePoint%22%3A%7B%22knowledgePoints%22%3A%5B%7B%22id%22%3A%2263%22%2C%22name%22%3A%22%22%2C%22isLeaf%22%3Afalse%2C%22enName%22%3A%22%22%7D%5D%7D%7D"

    attempts = 0
    while attempts < retries:
        problems_data = request_with_cookies(api_url, cookies_file_path)

        if problems_data:
            problems = problems_data.get('problems', [])
            for problem in problems:
                try:
                    # 打印调试信息
                    print(f"抓取题目: {problem.get('id')}, 难度: {problem.get('difficulty')}, 知识点: {problem.get('knowledgePoints')}")
                except Exception as e:
                    print(f"题目 {problem.get('id')} 处理出错，跳过此题。错误信息: {str(e)}")
                    continue
            return problems
        else:
            attempts += 1
            print(f"请求失败，状态码: 500，重试 {attempts}/{retries}")
            time.sleep(2)  # 等待 2 秒再重试

    print(f"题目页面 {page} 多次重试失败，跳过该页")
    return []

# 保存数据到 JSON 文件
def save_data_to_json(data, file_name='problems_data.json'):
    with open(file_name, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print(f"数据已保存到 {file_name}")

# 记录抓取进度到文件
def save_progress(progress, file_name='progress.txt'):
    with open(file_name, 'w') as f:
        f.write(str(progress))
    print(f"抓取进度已保存到 {file_name}")

# 读取上次抓取的进度
def load_progress(file_name='progress.txt'):
    try:
        with open(file_name, 'r') as f:
            progress = int(f.read().strip())
            return progress
    except FileNotFoundError:
        return 0  # 如果文件不存在，从0开始

# 递归抓取所有页数的数据，并从上次中断的地方继续
def fetch_all_problems(cookies_file_path='cookies.pkl'):
    page = load_progress()  # 从记录的页码开始
    limit = 30
    all_problems = []

    while True:
        problems = fetch_problems(page, limit, cookies_file_path)
        if problems:  # 如果有成功抓取到题目
            all_problems.extend(problems)
            page += 1  # 继续抓取下一页
            save_progress(page)  # 保存抓取进度
        else:
            # 如果没有题目返回或者整个页面跳过，继续下一页
            page += 1
            save_progress(page)  # 保存进度避免重复抓取
        # time.sleep(1)  # 增加延时，避免请求过快

    return all_problems

# 使用示例
if __name__ == "__main__":
    all_problems_data = fetch_all_problems()

    if all_problems_data:
        save_data_to_json(all_problems_data)
```
